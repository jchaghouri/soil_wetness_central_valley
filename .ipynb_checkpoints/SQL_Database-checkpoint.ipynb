```python
import pandas as pd
alamedadf = pd.read_csv('Data/Alameda County/AlmedaMIDPoint.csv')
merceddf = pd.read_csv('Data/Merced County/MercedMIDPoint.csv')
sacdf = pd.read_csv('Data/Sac County/SacramentoMIDPoint.csv')
sanjoaquindf = pd.read_csv('Data/San Joaqin County/SanJoaquinMIDPoint.csv')
solanodf = pd.read_csv('Data/Solano County/SolanoMIDPoint.csv')
stanislausdf = pd.read_csv('Data/Stanislaus County/StanislausMIDPoint.csv')
yolodf = pd.read_csv('Data/Yolo County/YoloMIDPoint.csv')

alamedadf['Name'] = 'Alameda'
alamedadf = alamedadf.loc[:,'YEAR':'Name']

merceddf['Name'] = 'Merced'
merceddf = merceddf.loc[:,'YEAR':'Name']

sacdf['Name'] = 'Sacramento'
sacdf = sacdf.loc[:,'YEAR':'Name']

sanjoaquindf['Name'] = 'San Joaquin'
sanjoaquindf = sanjoaquindf.loc[:,'YEAR':'Name']

solanodf['Name'] = 'Solano'
solanodf =solanodf.loc[:,'YEAR':'Name']

stanislausdf['Name'] = 'Stanislaus'
stanislausdf = stanislausdf.loc[:,'YEAR':'Name']

yolodf['Name'] = 'Yolo'
yolodf = yolodf.loc[:,'YEAR':'Name']

counties = [alamedadf,merceddf,sacdf,sanjoaquindf,solanodf,stanislausdf,yolodf]

data = pd.concat(counties)
data
```

```python
import geopandas as gpd
def read_file(path):
    return gpd.read_file(path)
geodata = read_file('county_ca.geojson')
```

```python
geodata
```

```python
countydf = geodata.merge(data,left_on ='NAME', right_on = 'Name', how ='outer')
countydf = countydf[['COUNTYFP', 'Name']]

countydf = countydf.drop_duplicates(subset = ["COUNTYFP", "Name"],keep='first')
countydf = countydf.dropna()
countydf

```

```python
import pandas as pd
import sqlite3

conn = sqlite3.connect('soil_database')
c = conn.cursor()

c.execute('CREATE TABLE IF NOT EXISTS counties (countyfp number, Name text)')
conn.commit()

c.execute('CREATE TABLE IF NOT EXISTS soil_wetness (YEAR number, JAN number, FEB number, MAR number, APR number, MAY number, JUN number, JUL number, AUG number, SEP number, OCT number, NOV number, DEC number, ANN number, Name text)')
conn.commit()

data.to_sql('soil_wetness', conn, if_exists='replace', index = False)


#c.execute('CREATE TABLE IF NOT EXISTS counties (countyfp number, Name text,PRIMARY_KEY(countyfp))')
#conn.commit()



countydf.to_sql('counties', conn, if_exists='replace', index = False)
```
