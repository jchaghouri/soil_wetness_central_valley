```python
import sqlite3
import pandas as pd
import geopandas as gpd
import matplotlib.pyplot as plt 
import csv
import pandas as pd 
from numpy import loadtxt
import streamlit as st
import plotly.figure_factory as ff
import matplotlib.pyplot as plt
import numpy as np
import sys
from matplotlib import colors #import 
import scipy.optimize as optimize
```

```python
conn = sqlite3.connect('soil_database') 
c = conn.cursor()

c.execute('''
          SELECT c.county_fips_id,c.county_name, s.value,s.years
          FROM soil_wetness s
          inner join counties c on s.county_fips_id = c.county_fips_id
          where years = "2011" 
          
          ''')

Info = pd.DataFrame(c.fetchall(),columns=['county_fips_id','county_name','value','years'])
c.execute('''
          SELECT c.county_fips_id,c.county_name, s.value,s.years
          FROM soil_wetness s
          inner join counties c on s.county_fips_id = c.county_fips_id
          where county_name = "Alameda" 
          
          ''')

AlamedaDecade = pd.DataFrame(c.fetchall(),columns=['county_fips_id','county_name','value','years'])

c.execute('''
          SELECT c.county_fips_id,c.county_name, s.value,s.years
          FROM soil_wetness s
          inner join counties c on s.county_fips_id = c.county_fips_id
          where county_name = "Merced" 
          
          ''')

MercedDecade = pd.DataFrame(c.fetchall(),columns=['county_fips_id','county_name','value','years'])

c.execute('''
          SELECT c.county_fips_id,c.county_name, s.value,s.years
          FROM soil_wetness s
          inner join counties c on s.county_fips_id = c.county_fips_id
          where county_name = "Sacramento" 
          
          ''')

SacramentoDecade = pd.DataFrame(c.fetchall(),columns=['county_fips_id','county_name','value','years'])

c.execute('''
          SELECT c.county_fips_id,c.county_name, s.value,s.years
          FROM soil_wetness s
          inner join counties c on s.county_fips_id = c.county_fips_id
          where county_name = "San Joaquin" 
          
          ''')

SanJDecade = pd.DataFrame(c.fetchall(),columns=['county_fips_id','county_name','value','years'])

c.execute('''
          SELECT c.county_fips_id,c.county_name, s.value,s.years
          FROM soil_wetness s
          inner join counties c on s.county_fips_id = c.county_fips_id
          where county_name = "Solano" 
          
          ''')

SolanoDecade = pd.DataFrame(c.fetchall(),columns=['county_fips_id','county_name','value','years'])

c.execute('''
          SELECT c.county_fips_id,c.county_name, s.value,s.years
          FROM soil_wetness s
          inner join counties c on s.county_fips_id = c.county_fips_id
          where county_name = "Stanislaus"
          
          ''')

StanislausDecade = pd.DataFrame(c.fetchall(),columns=['county_fips_id','county_name','value','years'])

c.execute('''
          SELECT c.county_fips_id,c.county_name, s.value,s.years
          FROM soil_wetness s
          inner join counties c on s.county_fips_id = c.county_fips_id
          where county_name = "Yolo" 
          
          ''')

YoloDecade = pd.DataFrame(c.fetchall(),columns=['county_fips_id','county_name','value','years'])
```

```python
fig, axs = plt.subplots(2, 4,figsize = (15,15))
axs[0, 0].bar(AlamedaDecade["years"],AlamedaDecade["value"])
axs[0, 0].set_title('Alameda Yearly Average')

axs[0, 1].bar(MercedDecade["years"],MercedDecade["value"])
axs[0, 1].set_title('Merced Yearly Average')

axs[0, 2].bar(SacramentoDecade["years"],SacramentoDecade["value"])
axs[0, 2].set_title('Sacramento Yearly Average')

axs[0, 3].bar(SanJDecade["years"],SanJDecade["value"])
axs[0, 3].set_title('San Joaquin Yearly Average')

axs[1, 0].bar(SolanoDecade["years"],SolanoDecade["value"])
axs[1, 0].set_title('Solano Yearly Average')

axs[1, 1].bar(StanislausDecade["years"],StanislausDecade["value"])
axs[1, 1].set_title('Stanislaus Yearly Average  ')

axs[1, 2].bar(YoloDecade["years"],YoloDecade["value"])
axs[1, 2].set_title('Yolo Yearly Average ')

for ax in axs.flat:
    ax.set(xlabel='Year', ylabel='Averaged Soil Wetness')
    ax.set(ylim = (0,0.8))
```
